# Importing libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
columns = [
    "CRIM",
    "ZN",
    "INDUS",
    "CHAS",
    "NOX",
    "RM",
    "AGE",
    "DIS",
    "RAD",
    "TAX",
    "PTRATIO",
    "B",
    "LSTAT",
    "MEDV",
]
```

```python
df = pd.read_csv("housing.csv", header=None, delimiter="\s+", names=columns)
df.head(10)
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
corr_matrix = df.corr()
corr_matrix
```

```python
sns.pairplot(df)
```

```python
target_corr = corr_matrix["MEDV"].abs().sort_values(ascending=False)
target_corr
```

```python
X = df.iloc[:, :-1]  # all columns except last
Y = df.iloc[:, -1]  # last column only
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
print(X_train.shape, X_test.shape)
print(Y_train.shape, Y_test.shape)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
import pickle

pickle.dump(scaler, open("scaling.pkl", "wb"))
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
model.get_params()
```

```python
model_op = model.predict(X_test)
```

```python
plt.scatter(Y_test, model_op)
```

```python
pickle.dump(model, open("regmodel.pkl", "wb"))
```

```python
# Load saved model and scaler
model = pickle.load(open("regmodel.pkl", "rb"))

scaler = pickle.load(open("scaling.pkl", "rb"))

# New data
X_new = pd.DataFrame(
    {
        "CRIM": [1000],
        "ZN": [10],
        "INDUS": [1],
        "CHAS": [2],
        "NOX": [100],
        "RM": [500],
        "AGE": [12],
        "DIS": [100],
        "RAD": [200],
        "TAX": [0],
        "PTRATIO": [10],
        "B": [100],
        "LSTAT": [10],
    }
)

# Scale & Predict
X_new_scaled = scaler.transform(X_new)
y_pred = model.predict(X_new_scaled)
print("Prediction:", y_pred[0])
```
